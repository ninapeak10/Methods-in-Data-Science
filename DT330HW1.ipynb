```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns
```

```python
#a)
college= pd.read_csv('College.csv') 
print(college.head(5))
```

```python
#b)
college2 = pd.read_csv('College.csv', index_col=0) 
college2.head(5)
```

```python
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1) 
college3 = college3.set_index('College')
college= college3
college.head(5)
```

```python
#c)
college.describe()
```

```python
#d)
columns= ['Top10perc', 'Apps', 'Enroll']
scatter_matrix(college[columns],figsize=(5, 5))
plt.show
```

```python
#e)
college.boxplot(column='Outstate', by= 'Private', grid= False)
plt.title('Outstate Tuition by Private/Public')
plt.suptitle('')
plt.show()
```

```python
#f)
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
college.boxplot(column='Outstate', by='Elite', grid=False)
plt.title('Boxplot of Outstate Tuition by Elite Status')
plt.suptitle('')
plt.xlabel('Elite Status')
plt.ylabel('Outstate Tuition')
plt.show()
```

```python
#g)
fig, axes= plt.subplots(2, 2, figsize=(10, 10))
college['Apps'].plot.hist(bins=30, ax= axes[0, 0], title= 'Applications')
college['Enroll'].plot.hist(bins= 20, ax= axes[0, 1], title= 'Enrollments')
college['Outstate'].plot.hist(bins= 25, ax= axes[1, 0], title= 'Out of State Tuition')
college['Room.Board'].plot.hist(bins= 15, ax= axes[1, 1], title= 'Room and Board')
plt.tight_layout()
```

h) There is a clear increase in out of state tuition of by private colleges compared to non-private. Private average cost is about the same amount as of the max tuition for non private being around $11,250, and the average non-private tuition is around $6,250.

There is also a high increase in out of state tuition for elite status students being $17,500, compared to non-elite status average being just less than $10,000.

There is the highest frequency for students paying room and board around $4,000. 


```python
#a)
Boston= pd.read_csv('Boston (1).csv', index_col= 0) 
print(Boston.head(5))
```

```python
#b)
num_rows, num_columns = Boston.shape


print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}')
```

Each row represents an individual observation or record for the boston area.

The colums represent a different statistic relating to the boston area. Like crime, indrustry, tax, or age. 

```python
#c)
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.scatterplot(x='rm', y='medv', data=Boston)
plt.title('Average Rooms vs. Median Home Value')

plt.subplot(2, 2, 2)
sns.scatterplot(x='lstat', y='medv', data=Boston)
plt.title('Lower Status % vs. Median Home Value')

plt.subplot(2, 2, 3)
sns.scatterplot(x='crim', y='medv', data=Boston)
plt.title('Crime Rate vs. Median Home Value')

plt.subplot(2, 2, 4)
sns.scatterplot(x='nox', y='medv', data=Boston)
plt.title('Nitric Oxides vs. Median Home Value')

plt.tight_layout()
plt.show()
```

There are some trends in the top two scatterplots and could show some correlation in each. In crime rate vs median home value you can see average low crime rate but there are some increases for lower home value and also an increase in the highest median home value crime rates. Then there isn't much of a clear relationship between Nitric Oxides vs. Median home value besides some common nox values shown often by the vertical trends. 

```python
#d)
plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
sns.scatterplot(x='medv', y='crim', data=Boston)
plt.title('Median Home Value vs. Crime Rate')

plt.subplot(3, 2, 2)
sns.scatterplot(x='lstat', y='crim', data=Boston)
plt.title('Lower Status % vs. Crime Rate')

plt.subplot(3, 2, 3)
sns.scatterplot(x='dis', y='crim', data=Boston)
plt.title('Distance to Employment Centers vs. Crime Rate')

plt.subplot(3, 2, 4)
sns.scatterplot(x='tax', y='crim', data=Boston)
plt.title('Tax Rate vs. Crime Rate')

plt.subplot(3, 2, 5)
sns.scatterplot(x='ptratio', y='crim', data=Boston)
plt.title('Pupil-Teacher Ratio vs. Crime Rate')

plt.tight_layout()
plt.show()
```

There is a trend for a higher crime rate on the lower side of median home value and a big jump for the highest median home value. 

The lower the distance to employment center has a higher crime rate.

There is a wide display of crime rate for tax rate around 660. 

There is a surge in the crime rate at the 20:1 pupil teacher ratio. 


```python
#e)
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(y='crim', data=Boston)
plt.title('Boxplot of Crime Rates')

plt.subplot(2, 2, 2)
sns.boxplot(y='tax', data=Boston)
plt.title('Boxplot of Tax Rates')

plt.subplot(2, 2, 3)
sns.boxplot(y='ptratio', data=Boston)
plt.title('Boxplot of Pupil-Teacher Ratios')

plt.tight_layout()
plt.show()
```

The average of crime rate is 0 but there are some outliers with a higher crime rate.
There is a wide spread of tax rates and no outliers shown. 
There is a standard of around 19 pupils per teacher ratio but there are some outliers havign less than 14.
